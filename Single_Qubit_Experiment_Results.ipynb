Octave outputs from configuration. We have 4 elements: a voltage source, a qubit drive, a resonator drive, and a multi qubit drive.

* The resonator drive is RF2

* The multiqubit drive is RF3 (note that this hasn't been setup)

* the voltage source comes from the OPX (note that this hasn't been setup)

```python
# Allows us to temporarily ignore
# import warnings
# warnings.filterwarnings("ignore", category=DeprecationWarning) 

#Experiment config (stores parameters for the experiment we are doing)

with open("experiments.json", "r") as f:
    experiment_config = json.load(f)
OPXconfig = initialConfiguration
# Creates object we use to do our experiments (avoid creating multiple times to 
# avoid calibration)
nanoTest = Single_Qubit_Experiments()
```

### Set Qubit/Resonator LO and IF Frequency

```python
#Set LO Frequency and IF frequency of resonator
ResLOFrequency = 7.0e9
ResIFFrequency = 200e6

nanoTest.changeLOfrequency('resonator', ResLOFrequency)
nanoTest.changeIFfrequency('resonator', ResIFFrequency)

#Set LO Frequency and IF frequency of qubit
QubitLOFrequency = 6.3e9
QubitIFFrequency = 270e6

nanoTest.changeLOfrequency('qubit', QubitLOFrequency)
nanoTest.changeIFfrequency('qubit', QubitIFFrequency)

```

### Start Quantum Machine

```python
nanoTest.start_QM_octave()
```

### Calibrate Elements

```python
# Note: will calibrate each LO frequency with each IF frequency one-to-one

# List of resonator LO and IF frequencies to calibrate: 
ResLOFrequencies= [ResLOFrequency]
ResIFFrequencies = [ResIFFrequency]

# List of qubit LO and IF frequencies to calibrate: 
QubitLOFrequencies= [QubitLOFrequency]
QubitIFFrequencies = [QubitIFFrequency]

nanoTest.editCalibrationParams("resonator", "calibration_params.json", 
    LOFrequencies= [ResLOFrequency],
    IFFrequencies= [ResIFFrequency],
)
nanoTest.editCalibrationParams("qubit", "calibration_params.json", 
   LOFrequencies= [QubitLOFrequency],
    IFFrequencies= [QubitIFFrequency],
)

nanoTest.calibrate()
```
